```r
library(igraph)
```

```r
probability = c(0.003, 0.004, 0.01, 0.05, 0.1)
graphs = list()
for (p in probability)
{
    g = sample_gnp(1000, p, directed = FALSE)
    graphs[[length(graphs) + 1]] = g
    mean_p=mean(degree(g))
    cat ("\nMean of the degree distributions is: ", mean_p, "with p =", p)
    var_p=var(degree(g))
    cat ("\nVariance of the degree distributions is: ", var_p, "with p =", p)
}
for (p in probability)
{
    g = sample_gnp(1000, p, directed = FALSE)
    plot(degree.distribution(g),main=paste("Degree Distribution for p = ", p),xlab="Degree",ylab="Probability")
}
```

(b)

```r
for (i in 1:length(probability))
{
    p = probability[i]
    g = graphs[[i]]
    connected_number = 0
    cat("\nFor p = ", p)
    for (i in 1:1000)
    {
        g = sample_gnp(1000, p, directed = FALSE)
        if (is.connected(g))
            connected_number = connected_number + 1
    }
    cat("\nProbability that the network is connected is", connected_number / 1000)   
    if (is.connected(g))
        cat('\nGraph connected.')
    else
        cat('\nGraph not connected.')
    g_components = clusters(g)
    ix = which.max(g_components$csize) 
    g_giant = induced.subgraph(g, which(g_components$membership == ix)) 
    cat("\nNumber of nodes of GCC:", vcount(g_giant))
    cur_diameter = diameter(g_giant, directed = FALSE)
    cat("\nThe Diameter of GCC:", cur_diameter,"\n")
}
```

(c)

```r
library(igraph)

n = 1000 # of times to sweep over values of p (to be generated)
rand_net = 100
O = 0.1
p = O*(1/n)

gcc_vals = list() # nonlinear function of p
p_vals = list()

range_index = seq(0.1, 1.5, 0.1)
rand_net_index = seq(1, rand_net, 1)
for (O in range_index){
    norm_gcc = 0
    p_log = O*log(n)/n 
    for(j in rand_net_index){
        giant = erdos.renyi.game(n, p_log) # defaults to false
        giant_clust_comp = clusters(giant)
        norm_gcc = max(giant_clust_comp$csize) + norm_gcc
    }
    norm_gcc = (norm_gcc/rand_net)*(1/n)
    gcc_vals = c(gcc_vals, list(norm_gcc))
    p_vals = c(p_vals, list(p_log))
}
    
p_log
plot(p_vals, gcc_vals, main = "Normalized GCC Sizes vs p", xlab = "p", ylab = "Normalized GCC Sizes")
```

i

```r
c = 0.5
n = seq(100,10000,100)
p = c/n
expected_gcc_sizes = c()
Expected_gcc_size <- function(n,p)
{
  expected_gcc_size = 0
  gcc_size = 0
  for (i in 1:100)
  {
    g = sample_gnp(n, p, directed = FALSE)
    g.components = clusters(g)
    ix = which.max(g.components$csize)
    g.giant = induced.subgraph(g, which(g.components$membership == ix))
    gcc_size = gcc_size + length(V(g.giant))
  }
  expected_gcc_size=gcc_size/100
  return(expected_gcc_size) 
}
for (i in 1:length(n))
  expected_gcc_sizes = c(expected_gcc_sizes, Expected_gcc_size(n[i],p[i]))
plot(n, expected_gcc_sizes, main = "Expected Size of the GCC of ER networks c = 0.5", xlab = "nodes number", ylab ="Expexted size of the GCC")
```

(ii)

```r
c = 1
p=c/n
expected_gcc_sizes = c()
for (i in 1:length(n))
  expected_gcc_sizes = c(expected_gcc_sizes, Expected_gcc_size(n[i],p[i]))
plot(n, expected_gcc_sizes, main = 'Expected Size of the GCC of ER networks c = 1', xlab = 'nodes number', ylab ='Expexted size of the GCC')
```

```r
c = 1.1
p=c/n
expected_gcc_sizes1 = c()
for (i in 1:length(n))
  expected_gcc_sizes1 = c(expected_gcc_sizes1, Expected_gcc_size(n[i],p[i]))
c = 1.2
p=c/n
expected_gcc_sizes2 = c()
for (i in 1:length(n))
  expected_gcc_sizes2 = c(expected_gcc_sizes2, Expected_gcc_size(n[i],p[i]))
c = 1.3
p=c/n
expected_gcc_sizes3 = c()
for (i in 1:length(n))
  expected_gcc_sizes3 = c(expected_gcc_sizes3, Expected_gcc_size(n[i],p[i]))
plot(n, expected_gcc_sizes1,main = 'Expected Size of the GCC of ER networks with c = 1.1, 1.2, 1.3', xlab = 'nodes number', ylab ='Expexted size of the GCC')
points(n,expected_gcc_sizes2,col="red")
points(n,expected_gcc_sizes3,col="blue")
legend(100, 1700, legend=c("c = 1.1", "c = 1.2", "c = 1.3"), col=c("black", "red", "blue"), cex=0.8, pch=c(1,1,1))
```
