```r
library(igraph)
```

```r
fb_net=read_graph("facebook_combined.txt", format='ncol', directed=FALSE)
net=graph_from_edgelist(as_edgelist(fb_net), directed=FALSE)
```

```r
cat(paste("Facebook network connected:", is_connected(net)))
```

# Question 2

```r
cat("Network diameter:", diameter(net))
```

# Question 3 

```r
plot(degree_distribution(fb_net), main="Degree distribution of network", xlab = "Degree", ylab = "Probability")
```

```r
cat("Average network degree:", mean(degree(net)))
```

# Question 4

```r
log_degree = log(seq(0:max(degree(fb_net))))
log_distribution = log(degree.distribution(fb_net))
log_degree = log_degree[-which(is.infinite(log_distribution))]
log_distribution = log_distribution[-which(is.infinite(log_distribution))]
plot(log_degree, log_distribution, main="Degree Distribution of the Network (log-log)",xlab="Degree",ylab="Probability")
abline(lm(log_distribution~log_degree))
```

# Question 8

```r
core_nodes = c()
for (i in degree(fb_net))
{
    if (i > 200)
    {
        core_nodes = c(core_nodes, i)
    }
}
cat("The number of core nodes: ", length(core_nodes))
cat("\nAverage degree: ", mean(core_nodes))
```

# Question 9

```r
node_ID = c(1,108,349,484,1087)
networks = make_ego_graph(fb_net, 1, nodes=V(fb_net)[node_ID])
for (i in 1:5) {
    cat("\nNode ID:", node_ID[i])
    fastgreedy= cluster_fast_greedy(networks[[i]])
    cat("\nFast Greedy modularity:", modularity(fastgreedy))
    plot(networks[[i]], mark.groups=groups(fastgreedy), edge.arrow.size=.5, 
         vertex.color=(fastgreedy$membership), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Fast-Greedy, Node ID=",node_ID[i],collapse=""))
    
    edgebet = cluster_edge_betweenness(networks[[i]])
    cat("\nEdge Betweenness modularity:", modularity(edgebet))
    plot(networks[[i]], mark.groups=groups(edgebet), edge.arrow.size=.5, 
     vertex.color=(edgebet$membership ), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Edge-Betweenness, Node ID=",node_ID[i],collapse=""))
    
    infomap = cluster_infomap(networks[[i]])
    cat("\nInfomap modularity:", modularity(infomap))
    plot(networks[[i]], mark.groups=groups(infomap), edge.arrow.size=.5, 
     vertex.color=(infomap$membership), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Infomap, Node ID=",node_ID[i],collapse=""))
}
```

# Question 10

```r
node_ID = c(1,108,349,484,1087)
for (i in 1:5) {
    networks1 = induced_subgraph(fb_net, neighbors(fb_net, node_ID[i]))
    cat("\nNode ID:", node_ID[i])
    fastgreedy= cluster_fast_greedy(networks1)
    cat("\nFast Greedy modularity:", modularity(fastgreedy))
    plot(networks1, mark.groups=groups(fastgreedy), edge.arrow.size=.5, 
         vertex.color=(fastgreedy$membership), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Fast-Greedy, Node ID=",node_ID[i],collapse=""))
    
    edgebet = cluster_edge_betweenness(networks1)
    cat("\nEdge Betweenness modularity:", modularity(edgebet))
    plot(networks1, mark.groups=groups(edgebet), edge.arrow.size=.5, 
     vertex.color=(edgebet$membership ), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Edge-Betweenness, Node ID=",node_ID[i],collapse=""))
    
    infomap = cluster_infomap(networks1)
    cat("\nInfomap:", modularity(infomap))
    plot(networks1, mark.groups=groups(infomap), edge.arrow.size=.5, 
     vertex.color=(infomap$membership), vertex.size=5, vertex.label="", 
    main=paste("Community Structure using Infomap, Node ID=",node_ID[i],collapse=""))
}
```
