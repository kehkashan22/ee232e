```r
library(igraph)
```

```r
fb_net=read_graph("facebook_combined.txt", format='ncol', directed=FALSE)
net=graph_from_edgelist(as_edgelist(fb_net), directed=FALSE)
```

```r
cat(paste("Facebook network connected:", is_connected(net)))
```

# Question 2

```r
cat("Network diameter:", diameter(net))
```

# Question 3 

```r
plot(degree_distribution(fb_net), main="Degree distribution of network", xlab = "Degree", ylab = "Probability")
```

```r
cat("Average network degree:", mean(degree(net)))
```

# Question 4

```r
log_degree = log(seq(0:max(degree(fb_net))))
log_distribution = log(degree.distribution(fb_net))
log_degree = log_degree[-which(is.infinite(log_distribution))]
log_distribution = log_distribution[-which(is.infinite(log_distribution))]
plot(log_degree, log_distribution, main="Degree Distribution of the Network (log-log)",xlab="Degree",ylab="Probability")
abline(lm(log_distribution~log_degree))
```
